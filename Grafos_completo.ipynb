### ***Algoritmo Djakstra Modificado***

```python
import sys

def dijkstra(graph, start):
    # Inicializa las distancias con infinito y el nodo de inicio con distancia cero
    distances = {node: float('inf') for node in graph}
    distances[start] = 0

    # Conjunto de nodos no visitados
    unvisited_nodes = list(graph)

    # Registro de caminos
    paths = {node: [] for node in graph}

    while unvisited_nodes:
        # Selecciona el nodo no visitado con la distancia más corta
        current_node = min(unvisited_nodes, key=lambda node: distances[node])
        unvisited_nodes.remove(current_node)

        # Calcula las distancias tentativas desde el nodo actual a los nodos vecinos
        for neighbor, weight in graph[current_node].items():
            tentative_distance = distances[current_node] + weight
            if tentative_distance < distances[neighbor]:
                distances[neighbor] = tentative_distance
                # Actualiza el camino hacia el vecino
                paths[neighbor] = paths[current_node] + [current_node]

    return distances, paths

# Ejemplo de uso
if __name__ == "__main__":
    graph = {
        'A': {'B': 1, 'C': 4},
        'B': {'A': 1, 'C': 2, 'D': 5},
        'C': {'A': 4, 'B': 2, 'D': 1},
        'D': {'B': 5, 'C': 1}


    }
    start_node = 'A'

    shortest_distances, paths = dijkstra(graph, start_node)

    print("Distancias más cortas desde el nodo de inicio (", start_node, "):")
    for node, distance in shortest_distances.items():
        print(node, ":", distance)

    print("\nCaminos más cortos desde el nodo de inicio (", start_node, "):")
    for node, path in paths.items():
        print(node, ":", [start_node] + path)
```

### ***Algoritmo edge_betweenness_centrality***

```python
import networkx as nx

# Crear un grafo no dirigido y ponderado
G = nx.Graph()

# Agregar nodos
G.add_nodes_from(['A', 'B', 'C', 'D', 'E'])

# Agregar aristas con pesos
edges_with_weights = [('A', 'B', 2), ('A', 'C', 1), ('B', 'D', 3), ('C', 'D', 4), ('C', 'E', 2), ('D', 'E', 5)]

for edge in edges_with_weights:
    G.add_edge(edge[0], edge[1], weight=edge[2])

# Calcular la centralidad de intermediación de aristas
edge_betweenness = nx.edge_betweenness_centrality(G, normalized=False, weight='weight')

# Imprimir el resultado
for edge, centrality in edge_betweenness.items():
    print(f'Arista {edge}: Centralidad de Intermediación = {centrality}')
```

Para analizar los resultados, se puede observar las aristas con mayor valor de centralidad, ya que indican cuales son los "puentes" más importantes del grafo que conectan diferentes partes. Aristas con centralidad cercana a 0 tienen baja intermediación.


Los puntos de centralidad nos muestran los puntos calientes del grafo, como nuestra información importante se encuentra en las aristas, usamos edge para referirnos a las aristas. Al identificar los puntos calientes, sabremos las zonas más 'peligrosas' de nuestro problema, y a partir de estas, sacar conclusiones.

### ***Algortimo de Louvain Modularity***

```python
import networkx as nx
from networkx.algorithms import community

def louvain_modularity(G, weight='weight'):
    # Utiliza el algoritmo de Louvain implementado en NetworkX
    communities = community.greedy_modularity_communities(G, weight=weight)
    return list(communities)
```

### ***Algortimo de Kruskal***

```python
import networkx as nx

def kruskal(G):

  mst = nx.Graph()
  edges = G.edges(data='weight')
  edges = sorted(edges, key=lambda x: x[2])

  groups = [[n] for n in G.nodes()]

  for e in edges:
    u,v,w = e
    set_u = -1
    set_v = -1
    for i in range(len(groups)):
      if u in groups[i]:
        set_u = i
      if v in groups[i]:
        set_v = i
    if set_u != set_v:
      mst.add_edge(u,v,weight=w)
      groups[set_u] += groups[set_v]
      groups.pop(set_v)

  return mst
```

```python
G = nx.Graph()

# Agrega nodos
G.add_nodes_from([1, 2, 3, 4, 5, 6])

# Agrega aristas con peso custom
G.add_edge(1, 2, weight=0.2)
G.add_edge(1, 3, weight=0.5)
G.add_edge(2, 3, weight=1.0)
G.add_edge(2, 4, weight=0.7)
G.add_edge(3, 5, weight=0.3)
G.add_edge(4, 5, weight=0.1)
G.add_edge(4, 6, weight=0.9)

# Ejecuta kruskal
mst = kruskal(G)

# Imprima el arbol
print(mst.edges(data=True))
```

Este algoritmo va agregando aristas al árbol en orden ascendente de peso, uniendo componentes conexas sin crear ciclos.

## Análisis Temporal entre los Meses (Enero, Febrero, Marzo, Abril) con Análisis de Flujos Temporales

```python
import networkx as nx
import numpy as np

# Definición de nodos y aristas (tomando en cuenta tu descripción)
localidades = ['Antonio Nariño', 'Barrios Unidos', 'Bosa', 'Candelaria', 'Chapinero', 'Ciudad Bolivar', 'Engativá', 'Kennedy', 'Puente Aranda', 'Rafael Uribe',
               'Santa Fe', 'Suba', 'Teusaquillo', 'Tunjuelito', 'Usme']
momentos_del_dia = ['Madrugada', 'Mañana', 'Tarde', 'Noche']

# Crear los grafos para enero, febrero y marzo
grafo_enero = nx.Graph()
grafo_febrero = nx.Graph()
grafo_marzo = nx.Graph()

# Agregar nodos al grafo
for nodo in localidades + momentos_del_dia:
    grafo_enero.add_node(nodo)
    grafo_febrero.add_node(nodo)
    grafo_marzo.add_node(nodo)

# Definir aristas y pesos (simulados para el ejemplo)
def agregar_aristas_aleatorias(grafo):
    for loc in localidades:
        for momento in momentos_del_dia:
            peso = np.random.randint(1, 101)  # Peso aleatorio para la arista
            grafo.add_edge(loc, momento, weight=peso)

# Agregar aristas ponderadas a los grafos de enero, febrero y marzo
agregar_aristas_aleatorias(grafo_enero)
agregar_aristas_aleatorias(grafo_febrero)
agregar_aristas_aleatorias(grafo_marzo)

# Calcular las diferencias entre las matrices de adyacencia de los meses
diferencia_enero_febrero = np.abs(nx.to_numpy_array(grafo_enero) - nx.to_numpy_array(grafo_febrero))
diferencia_febrero_marzo = np.abs(nx.to_numpy_array(grafo_febrero) - nx.to_numpy_array(grafo_marzo))
diferencia_marzo_enero = np.abs(nx.to_numpy_array(grafo_marzo) - nx.to_numpy_array(grafo_enero))

# Calcular la diferencia total entre los meses (suma de las diferencias absolutas)
diferencia_total = diferencia_enero_febrero + diferencia_febrero_marzo + diferencia_marzo_enero

# Imprimir la matriz de diferencia total (puedes analizar esta matriz para entender las diferencias)
print("Matriz de Diferencia Total entre los Meses:")
print(diferencia_total)
```
