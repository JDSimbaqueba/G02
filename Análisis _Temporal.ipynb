# Análisis temporal
El análisis temporal en teoría de grafos se refiere al estudio de cómo cambian las relaciones y estructuras en un grafo a lo largo del tiempo. En el contexto de hurto a personas en Bogotá utilizando grafos, el análisis temporal puede ser útil para:


1.   Identificar patrones en la actividad delictiva a lo largo del tiempo. Esto se puede observar por medio de la diferencia de pesos de las aristas en el grafo mes a mes.
2.   Analizar la propagación de hurtos entre localidades: Mediante el análisis temporal se puede estudiar cómo los hurtos se propagan de una localidad a otra a través del tiempo (Esto se puede ver el archivo Análisis de Afectaciones por Cercanía).

En esta sección trabajaremos en identificar patrones en la actividad delictiva por medio de la diferencia de pesos en las aristas entre los meses de Enero a Abril.

# Análisis entre Enero y Febrero


## Diferencia entre Enero y Febrero
En esta sección trabajamos los datos insertados como matrices de adyacencia. Con esto podemoas hacer una diferencia de matrices de adyacencia y así ver que tanto cambian los datos de un mes a otro. En este caso que tanto aumentó o disminuyó el hurto a personas de Enero a Febrero.

```python
import networkx as nx
import numpy as np

# Matrices de adyacencia: En base a los datos insertados, hacemos una matriz de adyacencia para cada mes.
A_ene = nx.to_numpy_array(grafo_enero)
A_feb = nx.to_numpy_array(grafo_febrero)

''' A continuación puede imprimir las matrices de adyacencia de Enero y Febrero, para verififcar los datos de la Matriz diferencia.

print(f"Grafo de Adyacencia Enero: \n{A_ene}")
print()
print(f"Grafo de Adyacencia Febrero: \n{A_feb}")
print()
print()

'''

# Matriz diferencia: La resta de las matrices de adyacencia para ver la diferencia de enero a febrero
A_diff_ene_feb = A_feb - A_ene
print("Matriz de diferencias:")
print(A_diff_ene_feb)

# Indices de cambios
increase = np.where(A_diff_ene_feb > 0)
decrease = np.where(A_diff_ene_feb < 0)

# Asignar indices a variables
valid_inc = []
valid_dec = []

for ind in range(0, len(increase[0])):
  valid_inc.append((increase[0][ind], increase[1][ind]))


for ind in range(0, len(decrease[0])):
  valid_dec.append((decrease[0][ind], decrease[1][ind]))

# Mapear índices a nombres de nodos
idx_to_node_ene = {i: n for i, n in enumerate(grafo_enero.nodes)}
idx_to_node_feb = {i: n for i, n in enumerate(grafo_febrero.nodes)}

# Imprimir
print()
print()
print("Aristas que aumentaron en hurto a personas de enero a febrero: ")
printed_edges = set()  # Para rastrear las aristas ya impresas

for i, j in valid_inc:
    edge = (idx_to_node_ene[i], idx_to_node_feb[j])
    if edge not in printed_edges and (edge[1], edge[0]) not in printed_edges:
        print(f"Aumento en arista de {edge[0]} a {edge[1]}")
        printed_edges.add(edge)

print()
print("Aristas que disminuyeron en hurto a personas de enero a febrero:")
printed_edges = set()  # Reiniciar el conjunto para las aristas disminuidas

for i, j in valid_dec:
    edge = (idx_to_node_ene[i], idx_to_node_feb[j])
    if edge not in printed_edges and (edge[1], edge[0]) not in printed_edges:
        print(f"Disminución en arista de {edge[0]} a {edge[1]}")
        printed_edges.add(edge)
```

Al ejecutar la celda anterior, tenemos como resultado varias secciones:
1.   Matriz de Diferencia Ene - Feb: En esta se pueden ver los cambios de hurto a personas en cada localidad y cada momento del día. Los resultados negativos representa que en febrero hubo menos robos que en enero, mientras que si el resultado es positivo, significa que aumentaron los robos. Por otro lado, aquellos resultados con valor 0, son los que no tienen aristas (Ej: Entre Mañana y Madrugada no hay arista. Entre Chapinero y Candelaria no hay arista) o que no hubo cambio de incidencia de robos entre enero y febrero.

2.   Aumentos y Disminuciones: Es la traducción de la matriz de adyacencia, de esta manera podemos determinar de manera sencilla cuales aristas en especifico aumentaron y cuales disminuyeron.

# Análisis Tempral entre Enero, Febrero, Marzo y Abril

```python
import networkx as nx

grafos_temporales = [grafo_enero, grafo_febrero, grafo_marzo, grafo_abril]

# Inicializar variables para el total de cambios, la variación y el valor de cambio final
total_cambios = 0
variacion_robos = 0
cambio_final_por_arista = {}

# Iterar sobre los meses
for i in range(len(grafos_temporales) - 1):
    current_graph = grafos_temporales[i]
    next_graph = grafos_temporales[i + 1]

    # Calcular el total de cambios y la variación en los robos
    cambios_mes = {
        (u, v): next_graph[u][v]['weight'] - current_graph[u][v]['weight']
        for u, v in current_graph.edges
    }

    total_cambios += sum(cambios_mes.values())
    variacion_robos += abs(sum(cambios_mes.values()))

    # Actualizar el valor de cambio final por arista
    for edge, cambio in cambios_mes.items():
        if edge not in cambio_final_por_arista:
            cambio_final_por_arista[edge] = 0
        cambio_final_por_arista[edge] += cambio

# Imprimir resultados
print(f"Total de cambios en los robos durante los cuatro meses: {total_cambios}")
print(f"Variación en los robos durante los cuatro meses: {variacion_robos}")
print("\nValor de cambio final por arista:")
for edge, cambio_final in cambio_final_por_arista.items():
    print(f"Arista {edge}: {cambio_final}")

```

El resultado del análisis muestra que el "Total de cambios en los robos durante los cuatro meses" es 747. Este valor representa la suma global de las variaciones en los robos para todas las aristas durante el período de cuatro meses. Si es positivo, indica un aumento total en los robos; si es negativo, señala una disminución total. En este caso 747 evidencia que hubo indica un aumento total en los robos al final de los 4 meses.

Además, la "Variación en los robos durante los cuatro meses" es 3411. Esta variación se calcula como la suma de las diferencias absolutas entre los pesos de las aristas en meses consecutivos. Indica cuánto ha variado la cantidad total de robos en la red durante los cuatro meses. El valor 3411 refleja una mayor variabilidad en la cantidad de robos a lo largo de los 4 meses.

El análisis también presenta las aristas junto con su suma/variación total entre los meses. Este enfoque proporciona información para hacer una predicción sobre qué aristas pueden experimentar aumentos o disminuciones en el próximo mes.
