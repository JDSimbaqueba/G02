### ***Algortimo de Kruskal máximo***

```python
import networkx as nx

def kruskal_max_spanning_tree(G):

  max_tree = nx.Graph()
  for n in G.nodes():
    max_tree.add_node(n)

  edges = G.edges(data='weight')
  edges = sorted(edges, key=lambda x: x[2], reverse=True)

  for e in edges:
    u,v,w = e
    max_tree.add_edge(u, v, weight=w)
    if nx.is_tree(max_tree):
      break

  return max_tree
```

### ***Algoritmo Djakstra Modificado***

```python
import networkx as nx
import sys

def dijkstra(graph, start):
    # Inicializa las distancias con infinito y el nodo de inicio con distancia cero
    distances = {node: float('inf') for node in graph}
    distances[start] = 0

    # Conjunto de nodos no visitados
    unvisited_nodes = list(graph)

    # Registro de caminos
    paths = {node: [] for node in graph}

    while unvisited_nodes:
        # Selecciona el nodo no visitado con la distancia más corta
        current_node = min(unvisited_nodes, key=lambda node: distances[node])
        unvisited_nodes.remove(current_node)

        # Calcula las distancias tentativas desde el nodo actual a los nodos vecinos
        for neighbor, weight in graph[current_node].items():
            tentative_distance = distances[current_node] + weight['weight']
            if tentative_distance < distances[neighbor]:
                distances[neighbor] = tentative_distance
                # Actualiza el camino hacia el vecino
                paths[neighbor] = paths[current_node] + [current_node]

    return distances, paths

# Ejemplo de uso
if __name__ == "__main__":
    H = nx.Graph()

    H.add_edge('Suba', 'Engativa', weight=280)
    H.add_edge('Suba', 'Barrios Unidos', weight=191)

    H.add_edge('Engativa', 'Barrios Unidos', weight=193)
    H.add_edge('Engativa', 'Teusaquillo', weight=250)

    H.add_edge('Barrios Unidos', 'Chapinero', weight=162)
    H.add_edge('Barrios Unidos', 'Teusaquillo', weight=133)


    H.add_edge('Chapinero', 'Santa Fe', weight=176)
    H.add_edge('Chapinero', 'Teusaquillo', weight=190)


    H.add_edge('Teusaquillo', 'Puente Aranda', weight=155)


    H.add_edge('Santa Fe', 'Candelaria', weight=72)


    H.add_edge('Puente Aranda', 'Kennedy', weight=253)
    H.add_edge('Puente Aranda', 'Antonio Nariño', weight=106)
    H.add_edge('Puente Aranda', 'Tunjuelito', weight=94)

    H.add_edge('Kennedy', 'Tunjuelito', weight=199)
    H.add_edge('Kennedy', 'Bosa', weight=240)

    H.add_edge('Bosa', 'Ciudad Bolivar', weight=89)
    H.add_edge('Bosa', 'Tunjuelito', weight=81)


    H.add_edge('Antonio Nariño', 'Rafael Uribe', weight=83)
    H.add_edge('Antonio Nariño', 'Tunjuelito', weight=52)


    H.add_edge('Rafael Uribe', 'Usme', weight=96)
    H.add_edge('Rafael Uribe', 'Tunjuelito', weight=71)


    H.add_edge('Tunjuelito', 'Ciudad Bolivar', weight=48)

    start_node = 'Suba'

    # Obtener la representación del grafo como diccionario
    graph_dict = nx.to_dict_of_dicts(H)

    shortest_distances, paths = dijkstra(graph_dict, start_node)

    print("Distancias más cortas desde el nodo de inicio (", start_node, "):")
    for node, distance in shortest_distances.items():
        print(node, ":", distance)

    print("\nCaminos más cortos desde el nodo de inicio (", start_node, "):")
    for node, path in paths.items():
        print(node, ":", [start_node] + path)
```

```python
import networkx as nx
import sys

def dijkstra(graph, start):
    # Inicializa las distancias con infinito y el nodo de inicio con distancia cero
    distances = {node: float('inf') for node in graph}
    distances[start] = 0

    # Conjunto de nodos no visitados
    unvisited_nodes = list(graph)

    # Registro de caminos
    paths = {node: [] for node in graph}

    while unvisited_nodes:
        # Selecciona el nodo no visitado con la distancia más corta
        current_node = min(unvisited_nodes, key=lambda node: distances[node])
        unvisited_nodes.remove(current_node)

        # Calcula las distancias tentativas desde el nodo actual a los nodos vecinos
        for neighbor, weight in graph[current_node].items():
            tentative_distance = distances[current_node] + weight['weight']
            if tentative_distance < distances[neighbor]:
                distances[neighbor] = tentative_distance
                # Actualiza el camino hacia el vecino
                paths[neighbor] = paths[current_node] + [current_node]

    return distances, paths

# Ejemplo de uso
if __name__ == "__main__":
    H = nx.Graph()

    # Agregar datos al grafo H
    H.add_edge('Suba', 'Engativa', weight=318)
    H.add_edge('Suba', 'Barrios Unidos', weight=244)

    H.add_edge('Engativa', 'Barrios Unidos', weight=223)
    H.add_edge('Engativa', 'Teusaquillo', weight=273)


    H.add_edge('Barrios Unidos', 'Chapinero', weight=171)
    H.add_edge('Barrios Unidos', 'Teusaquillo', weight=160)


    H.add_edge('Chapinero', 'Santa Fe', weight=218)
    H.add_edge('Chapinero', 'Teusaquillo', weight=182)


    H.add_edge('Teusaquillo', 'Puente Aranda', weight=167)


    H.add_edge('Santa Fe', 'Candelaria', weight=136)


    H.add_edge('Puente Aranda', 'Kennedy', weight=269)
    H.add_edge('Puente Aranda', 'Antonio Nariño', weight=111)
    H.add_edge('Puente Aranda', 'Tunjuelito', weight=104)

    H.add_edge('Kennedy', 'Tunjuelito', weight=210)
    H.add_edge('Kennedy', 'Bosa', weight=251)

    H.add_edge('Bosa', 'Ciudad Bolivar', weight=88)
    H.add_edge('Bosa', 'Tunjuelito', weight=86)


    H.add_edge('Antonio Nariño', 'Rafael Uribe', weight=92)
    H.add_edge('Antonio Nariño', 'Tunjuelito', weight=51.5)


    H.add_edge('Rafael Uribe', 'Usme', weight=86)
    H.add_edge('Rafael Uribe', 'Tunjuelito', weight=85)


    H.add_edge('Tunjuelito', 'Ciudad Bolivar', weight=47)

    start_node = 'Candelaria'

    # Obtener la representación del grafo como diccionario
    graph_dict = nx.to_dict_of_dicts(H)

    shortest_distances, paths = dijkstra(graph_dict, start_node)

    print("Distancias más cortas desde el nodo de inicio (", start_node, "):")
    for node, distance in shortest_distances.items():
        print(node, ":", distance)

    print("\nCaminos más cortos desde el nodo de inicio (", start_node, "):")
    for node, path in paths.items():
        print(node, ":", [start_node] + path)
```
